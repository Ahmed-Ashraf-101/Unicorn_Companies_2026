```python
import requests

from bs4 import BeautifulSoup

import csv
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"}

page = requests.get("https://www.cbinsights.com/research-unicorn-companies", headers= headers)

print(page.status_code)
```

```python
def main(page):
    source = page.content
    soup = BeautifulSoup(source, 'lxml')
    table_data = []

    companies = soup.find_all('tr')
    for row in companies:
        cells = [td.get_text(strip= True) for td in row.find_all('td')]
        if cells:
            table_data.append(cells)

    headers = ["Company", "Valuation ($B)", "Date Joined", "Country", "City", "Industry", "Select Investors"]

    with open(r"D:\Career\Web Scraping\Unicorn_Companies\Unicorn_Companies.csv", "w", newline= '', encoding= 'utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(headers)
        writer.writerows(table_data)

    print("File Created")
```

```python
main(page)
```
